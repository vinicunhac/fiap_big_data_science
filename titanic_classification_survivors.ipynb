## 0. Libs

```python
import pandas as pd
import sklearn.preprocessing as skpp
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import recall_score, precision_score, accuracy_score
```

## 1. Loading datasets

```python
df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/dados_arvore_titanic.csv')
df.head()
```

## 2. Feature Engineering

```python
#Construção da dummy de gênero
dummies_sex = pd.get_dummies(df['Sex'])
df['SexFemale'] = dummies_sex['female']

dummies_class = pd.get_dummies(df['Pclass'], prefix='p')
df['p1'] = dummies_class['p_1']
df['p2'] = dummies_class['p_2']

age_med = df['Age'].mean()
df['Age'] = df['Age'].fillna(age_med)

df.head()
```

## 3. Sample train and test

```python
# Survived é a variável target e SexFemale é a feature
x_train, x_test, y_train, y_test = train_test_split(df[['SexFemale','p1', 'p2', 'Age']], df['Survived'], test_size=0.25, random_state=33)
```

## 4. Tree Models

```python
mod = tree.DecisionTreeClassifier()

mod.fit(x_train, y_train)
y_prev = mod.predict(x_test)
```

## 5. Confusion Matrix

```python
# Confusion Matrix
pd.crosstab(y_prev, y_test, margins=True)
```

## 6. Testing model

```python
# Métricas de Erro de previsão
print(accuracy_score(y_test, y_prev))
print(recall_score(y_test, y_prev))
print(precision_score(y_test, y_prev))
```
