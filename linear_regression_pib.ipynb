```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as dash
from google.colab import drive
import os
import warnings
import itertools
warnings.filterwarnings('ignore')
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/My Drive/Colab Notebooks/FIAP - Data Science/Pasta.xlsx', index_col=0)  
df.head()
```

```python
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
import statistics

schema_list = df.columns.tolist()
del schema_list[0]
final_list = []
for a in range(2,12):
  for i in itertools.combinations(schema_list, a):

    filter =  [(i[j]) for j in range(len(i))]
    
    model1=sm.OLS(df['Indice PIB'], df[filter])

    result=model1.fit()

    intern_list = [f'{filter}', statistics.mean([(result.pvalues[j]) for j in range(len(result.pvalues))]), result.rsquared,  statistics.mean([(result.bse[j]) for j in range(len(result.bse))])]

    final_list.append(intern_list)

```

```python
df_1 = pd.DataFrame(final_list, columns =['combination', 'pvalues', 'rsquared', 'stderror'])
```

```python
df_1.head()
```

```python
filter = df_1["pvalues"] <= 0.05
df_1.where(filter, inplace = True)
```
